```python
import json
import requests
import time
import pandas as pd
```

```python
with open("JSON_DATA/pro_players_list.json", "r") as read_file:
    players_list = json.load(read_file)
```

```python
for player in players_list:
    check = False
    while check == False:
        try:
            accaunt_id = str(player['account_id'])
            url = 'https://api.opendota.com/api/players/' + accaunt_id+ '/recentMatches'
            r = requests.get(url)
            with open('JSON_DATA/'+ accaunt_id + 'update.json', 'wb') as outfile:
                outfile.write(r.content)
            time.sleep(0.5)
            print('.', end=' ')
            check = True
            data = pd.read_json("JSON_DATA/88470.json")
            data_update = pd.read_json("JSON_DATA/88470update.json")
            data = pd.concat([data, data_update])
            data = data.reset_index(drop=True)
            data = data.sort_values(by=['match_id'], ascending=False)
            data = data.drop_duplicates()
            data = data.reset_index(drop=True)
            data.to_json("JSON_DATA/" + "88470" + ".json", orient = 'records')
        except: time.sleep(1)
```
